```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web
```

```python
style.use('ggplot')
```

```python
start = dt.datetime(2019,1,1)
end = dt.datetime.today()
```

```python
companies = ["msft", "amzn", "tsla", "dal", "aapl"]
```

```python
df = web.DataReader(companies, 'yahoo', start, end)
```

```python
df.to_csv('companies.csv')
```

```python
csv = pd.read_csv('companies.csv', parse_dates = True, index_col=0)
```

```python
df[['Close', 'Open', 'High', 'Low']].plot()
```

```python
apple = web.DataReader('AAPL', 'yahoo', start, end)
apple.to_csv('aapl.csv')
apple = pd.read_csv('aapl.csv', parse_dates = True, index_col=0)
apple['100ma'] = apple['Adj Close'].rolling(window=100, min_periods=0).mean()
apple.dropna(inplace=True)
apple[['Close', 'Open', 'High', 'Low', '100ma']].plot()
```

```python
microsoft = web.DataReader('msft', 'yahoo', start, end)
microsoft.to_csv('msft.csv')
microsoft = pd.read_csv('msft.csv', parse_dates = True, index_col=0)
microsoft['100ma'] = microsoft['Adj Close'].rolling(window=100, min_periods=0).mean()
microsoft.dropna(inplace=True)
microsoft[['Close', 'Open', 'High', 'Low', '100ma']].plot()
```

```python
delta = web.DataReader('DAL', 'yahoo', start, end)
delta.to_csv('dal.csv')
delta = pd.read_csv('dal.csv', parse_dates = True, index_col=0)
delta['100ma'] = delta['Adj Close'].rolling(window=100, min_periods=0).mean()
delta.dropna(inplace=True)
delta[['Close', 'Open', 'High', 'Low', '100ma']].plot()
```
